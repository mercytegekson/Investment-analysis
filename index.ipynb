# Task One : Data Overview

This dataset is a study on investment behavior and preferences among individuals. It contains demographic information, investment choices, motivations, and expectations regarding various investment avenues.

# Data Description

## Demographic Information:

Gender: Categorizes respondents by gender, allowing analysis of investment preferences by gender.

Age: Contains respondents' ages, enabling examination of how investment choices vary across age groups.

## Investment Avenues:

Investment_Avenues: Likely a general category or summary of investment types chosen.

Mutual_Funds: Indicates if/how much the person invests in mutual funds.

Equity_Market: Tracks investments directly in equities/stocks.

Debentures: Records investment in corporate bonds/debentures.

Government_Bonds: Shows investment in government securities.

Fixed_Deposits: Indicates money placed in bank fixed deposits.

PPF: Records investment in Public Provident Fund, a government savings scheme.

Gold: Tracks investment in gold (physical or paper gold).

Stock_Market: Similar to Equity_Market, possibly capturing overall stock market participation.

## Investment Decision Factors:

Factor: Likely captures the primary factor influencing investment decisions (e.g., risk, return, liquidity).

Objective: Records the financial objective behind investments (e.g., retirement, education).

Purpose: Similar to objective, may focus on specific goals for the investments.

Duration: Indicates the time horizon for which investments are planned.

Invest_Monitor: Shows how frequently or by what method investors track their investments.

Expect: Captures expected returns or performance expectations from investments.

Avenue: May represent the preferred or primary investment avenue.

## Survey Questions:

What are your savings objectives?: A direct question from the survey about financial goals.

Investment Reasoning:

Reason_Equity: Specific reasons for choosing equity investments.

Reason_Mutual: Rationale behind mutual fund investments.

Reason_Bonds: Reasoning for bond investments.

Reason_FD: Explanation for fixed deposit choices.

## Information Source:

Source: Indicates where respondents get their investment information (e.g., financial advisors, internet, family).

# Task 10: Correlation Analysis

```python
numerical_columns = data[['age', 'Mutual_Funds', 'Equity_Market', 'Debentures','Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']]
```

```python
# Correlation matrix
correlation_matrix = numerical_columns.corr()
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Investment Options')
plt.show()
```

Based on the heatmap, Equity Market and Mutual Funds show a strong positive correlation, indicating that investors aiming for growth often prefer both. Conversely, risk-averse instruments like PPF and Fixed Deposits are negatively correlated with equities and debentures, reflecting conservative investor preferences. Age positively correlates with debentures and equity markets, suggesting older individuals lean toward more mature investments. Mutual Funds and Gold exhibit a strong negative correlation, highlighting differing risk appetites. Overall, investors tend to group their choices into either high-risk/high-return or low-risk/stable return categories.
